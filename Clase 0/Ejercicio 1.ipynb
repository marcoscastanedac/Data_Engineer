```python
# Desvinculaciones por años de antiguedad
df_desvinculaciones_antiguedad = df_udm['antiguedad_ano'].value_counts().sort_index().reset_index()
df_desvinculaciones_antiguedad.columns = ['Antiguedad', 'Cantidad']
df_desvinculaciones_antiguedad.plot(kind='bar',  x='Antiguedad', y='Cantidad', title='Cantidad de Desvinculaciones por Antigüedad en Años', color='skyblue')
plt.xlabel('Antigüedad en Años')
plt.ylabel('Cantidad de Desvinculaciones')
plt.show()
```

**Se determina que ni cumpliendo 1 año en la empresa se genera una desvinculación**

```python
# Desvinculaciones por Manager
# Calculamos el número de despidos por manager y ordenamos los resultados
df_desvinculaciones_manager = df_udm['Nombre_y'].value_counts().reset_index()
df_desvinculaciones_manager.columns = ['Nombre_y', 'Cantidad_Despidos']
df_desvinculaciones_manager = df_desvinculaciones_manager.sort_values(by='Cantidad_Despidos', ascending=False)

# Obténemos el nombre del manager con más despidos
manager_mas_despidos = df_desvinculaciones_manager.iloc[0]['Nombre_y']
cantidad_mas_despidos = df_desvinculaciones_manager.iloc[0]['Cantidad_Despidos']

# Gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(df_desvinculaciones_manager['Nombre_y'], df_desvinculaciones_manager['Cantidad_Despidos'], color='skyblue')

# Resaltar al manager con más despidos
for bar in bars:
    if bar.get_height() == cantidad_mas_despidos:
        bar.set_color('red')  # Cambiar el color del bar del manager con más despidos

plt.title('Cantidad de Despidos por Manager')
plt.xlabel('Manager')
plt.ylabel('Cantidad de Despidos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Se determina que el manager Ricardo esta asociado al mayor numero de desvinculaciónes con 6**

```python
# Desvinculaciones por rango salarial
# Agrupamos por rango salarial ('desde' y 'hasta') y calcular el número de desvinculaciones
df_desvinculaciones_salario = df_udm.groupby(['desde', 'hasta']).size().reset_index(name='Cantidad')

# Ordenar los resultados por 'desde' (rango salarial)
df_desvinculaciones_salario = df_desvinculaciones_salario.sort_values(by='desde')

# Gráfico de barras de desvinculaciones por rango salarial
plt.figure(figsize=(10, 6))
plt.bar(df_desvinculaciones_salario.index, df_desvinculaciones_salario['Cantidad'], color='skyblue')
plt.title('Cantidad de Desvinculaciones por Rango Salarial')
plt.xlabel('Índice del Rango Salarial')
plt.ylabel('Cantidad de Desvinculaciones')
plt.xticks(df_desvinculaciones_salario.index, df_desvinculaciones_salario['desde'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Desvinculaciones por Grupo de Reclutamiento
df_desvinculaciones_reclutador = df_udm['Grupo_recruitment'].value_counts().sort_index().reset_index()
df_desvinculaciones_reclutador.columns = ['Grupo', 'Cantidad']
df_desvinculaciones_reclutador.plot(kind='bar',  x='Grupo', y='Cantidad', title='Cantidad de Desvinculaciones por Grupo de Reclutamiento', color='skyblue')
plt.xlabel('Grupo Recruiter')
plt.ylabel('Cantidad de Desvinculaciones')
plt.show()
```

```python
# Desvinculaciones por Tiempo de Reclutamiento
df_desvinculaciones_reclutador = df_udm['Tiempo_recruitment'].value_counts().sort_index().reset_index()
df_desvinculaciones_reclutador.columns = ['Tiempo Reclutamiento', 'Cantidad']
df_desvinculaciones_reclutador.plot(kind='bar',  x='Tiempo Reclutamiento', y='Cantidad', title='Cantidad de Desvinculaciones por Tiempo de Reclutamiento', color='skyblue')
plt.xlabel('Tiempo Reclutamiento')
plt.ylabel('Cantidad de Desvinculaciones')
plt.show()
```

```python
# Desvinculaciones por Curso
# Calculamos el número de desvinculaciones por cursos realizados y no realizados
df_desvinculaciones_curso = df_udm['Realizo_cursos'].value_counts().reset_index()
df_desvinculaciones_curso.columns = ['Realizo_cursos?', 'Cantidad']

# Gráfico de barras
plt.figure(figsize=(8,6))
bars = plt.bar(df_desvinculaciones_curso['Realizo_cursos?'], df_desvinculaciones_curso['Cantidad'], color=['skyblue', 'lightcoral'])

# Añadir etiquetas y título
plt.title('Cantidad de Desvinculaciones por Cursos Realizados')
plt.xlabel('¿Realizó cursos?')
plt.ylabel('Cantidad de Desvinculaciones')

# Mostrar el valor de cada barra
for bar in bars:
    yval = bar.get_height()
    # Mostramos el texto encima de las barras para mayor detalle
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')

plt.xticks(rotation=0)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

**Se determina que entre más cursos, más desvinculaciones existen**
